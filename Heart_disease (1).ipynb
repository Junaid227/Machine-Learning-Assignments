```python
X=df.drop(['target'],axis=1)
y=df['target']
```

```python
y.value_counts()
```

```python
X.describe()
```

Train,Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression(max_iter=1000)
```

```python
X_train
```

```python
model.fit(X_train,y_train)
```

```python
#predicting model
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
#classification report
print(classification_report(y_test,predictions))
```

```python
from collections import Counter
Counter(predictions)
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics
```

```python
#create decision tree classfier object
clf=DecisionTreeClassifier()
#train decision tree classifier
clf.fit(X_train,y_train)
#predict the response for dataset
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
#model accuracy testing
print('test accuracy:', sklearn.metrics.accuracy_score(y_test,y_pred))
```

```python
y_pred_train=clf.predict(X_train)
```

```python
print('train accuracy:', sklearn.metrics.accuracy_score(y_pred_train,y_train))
```

```python
#classificatin report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data=StringIO()
export_graphviz(clf,
                out_file=dot_data,
                filled=True,
                rounded=True,
                special_characters=True,
                feature_names=X.columns,
                class_names=['0','1'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('Heart Disease.png')
Image(graph.create_png())
```

```python
#create decision tree classifier object
clf= DecisionTreeClassifier(criterion='entropy',max_depth=3)

# Train Decision Tree Classifer
clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_train)

# # Train Decision Tree Classifer
# clf = clf.fit(X_train,y_train)


# Model Accuracy, how often is the classifier correct?
print("Train Accuracy:",sklearn.metrics.accuracy_score(y_train, y_pred))

##### Test###

y_pred_test = clf.predict(X_test)
# Model Accuracy
print("Test Accuracy:",sklearn.metrics.accuracy_score(y_test, y_pred_test))
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
accuracy_score(y_test,y_pred_test)#for decision tree after pruning
```
